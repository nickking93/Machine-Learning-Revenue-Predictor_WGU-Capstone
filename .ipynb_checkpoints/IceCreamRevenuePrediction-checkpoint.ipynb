<h2><center>Model Training</center></h2>

<h3>Linear Regression</h3>

We will employ a linear regression model to predict the revenue of Frosty Forecasts Ice Cream Co. based on the same data being displayed in the charts above. Linear regression aims to establish a linear relationship between temperature and the revenue, and will model this relationship as a straight line. The equation of the line is represented as $y = mx + b$, where $y$ is the predicted revenue, $x$ is the temperature, $m$ is the slope, and $b$ is the y-intercept. 

```python
# To view this code, click the three dots -->

# Linear regression
regression = LinearRegression()
regression.fit(x, y)

# Plot the line
plt.figure(figsize=(10,6))
plt.scatter(x,y, alpha=0.3)
plt.plot(x, regression.predict(x), color='red', linewidth=4)

plt.title('Ice Cream Revenue by Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Revenue $')
plt.ylim(10, 1000)
plt.xlim(0, 45.0)
plt.show()
```

##### <p align="justify">The chart above draws the the regression line on the previously shown scatterplot and illustrates the relationship between temperature and revenue. While the scatterplot displays individual data points with actual values, the overlaid regression line represents the model’s predicted revenue based on temperature, clearly showing the linear trend the model has identified. This visualization not only helps in understanding the strength and direction of the relationship but also provides a quick assessment of how well the linear regression model fits the data. By comparing the data points to the regression line, we can easily gauge the model's accuracy.</p>

<h3>Gradient Descent</h3>

<p align="justify">To optimize the parameters $m$ and $b$, we use gradient descent, an iterative optimization algorithm. Gradient descent minimizes the cost function, which measures the difference between the actual and predicted revenues. By iteratively adjusting $m$ and $b$ in the direction that reduces the cost function, gradient descent efficiently converges to the optimal values that best fit the historical data, thus improving the accuracy of the revenue predictions.</p>

```python
# To view this code, click the three dots -->

# Convert lists to NumPy arrays
x1 = np.array(temperature)
y1 = np.array(revenue)
m1 = 21.4436
b1 = 44.8312
# Add a column of ones to x1 to include the intercept term
x1_b = np.c_[x1, np.ones_like(x1)]

# Initialize parameters
m = m1  # Slope (coefficient for temperature)
b = b1  # Intercept
learning_rate = 0.01
precision = .01  # Precision for convergence
max_iterations = 5000000  # Maximum number of 

# Function to compute the cost (Mean Squared Error)
def compute_cost(x1, y1, m, b):
    predictions = x1 @ np.array([m, b])
    errors = predictions - y1
    return np.mean(errors ** 2)
    
# Set up the plot
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1, color='blue', label='Data Points', alpha=0.3)
line, = plt.plot(x1, x1 * m + b, color='red', label='Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Revenue')
plt.title('Gradient Descent Optimization of Linear Regression')
plt.legend()

# Gradient Descent with Convergence Criteria
previous_cost = float('inf')  # Initialize with a large 

for iteration in range(max_iterations):
    # Compute predictions
    predictions = x1_b @ np.array([m, b])
    
    # Compute gradients
    error = predictions - y1
    gradient_m = (2 / len(y1)) * np.sum(error * x1)
    gradient_b = (2 / len(y1)) * np.sum(error)
    
    # Update parameters
    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b
    
    # Compute the current cost
    current_cost = compute_cost(x1_b, y1, m, b)
    stepsize = abs(previous_cost - current_cost)
    
    # Check for convergence
    if stepsize < precision:
        print(f"Converged after {iteration+1} iterations.")
        break
    else:
        break
    
    # Update the previous cost
    previous_cost = current_cost
    
    # Update the regression line in the plot
    line.set_ydata(x1 * m + b)
    plt.pause(0.01)  # Pause to update the plot

print(f"{bcolors.BOLD}{bcolors.UNDERLINE}Starting slope (m): {bcolors.ENDC} {bcolors.OKGREEN}{m1}{bcolors.ENDC} ")
print(f"{bcolors.BOLD}{bcolors.UNDERLINE}Optimized slope (m): {bcolors.ENDC} {bcolors.OKGREEN}{m}{bcolors.ENDC} ")
print('')
print(f"{bcolors.BOLD}{bcolors.UNDERLINE}Starting intercept (b): {bcolors.ENDC} {bcolors.OKGREEN}{b1}{bcolors.ENDC}")
print(f"{bcolors.BOLD}{bcolors.UNDERLINE}Optimized intercept (b): {bcolors.ENDC} {bcolors.OKGREEN}{b}{bcolors.ENDC}")
```

<p align="justify">Despite the small changes observed in the optimized values of $m$ (slope) and $b$ (intercept), the process of applying gradient descent and visualizing the updates remains highly instructive and valuable. The initial regression line already provided extremely accurate predictions, indicating a strong fit to the data. However, the gradient descent optimization allowed for fine-tuning of the parameters, ensuring that even minor inaccuracies were minimized. The chart was updated to reflect these small adjustments and visually demonstrates the robustness and precision of the model. This iterative approach reinforces the reliability of the model's predictions and highlights the power of gradient descent in achieving optimal parameter values, even when starting from a nearly optimal state. Thus, the exercise underscores the importance of gradient descent in model optimization and provides a clear, visual confirmation of its effectiveness.</p>

<h2><center>Validation</center></h2>

```python
# Initialize k-fold cross-validation
x = DataFrame(data, columns=['Temperature'])
y = DataFrame(data, columns=['Ice Cream Profits'])
kf = KFold(n_splits=5, shuffle=True, random_state=1)  # 5-fold cross-validation
regression = LinearRegression()

# Lists to store performance metrics
r2_scores = []
mae_scores = []
rmse_scores = []

# Perform cross-validation
for train_index, test_index in kf.split(x):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit the model
    regression.fit(x_train, y_train)
    
    # Make predictions
    y_pred = regression.predict(x_test)
    
    # Calculate performance metrics
    r2_scores.append(r2_score(y_test, y_pred))
    mae_scores.append(np.mean(np.abs(y_test - y_pred)))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))

# Calculate average performance metrics
avg_r2 = np.mean(r2_scores)
avg_mae = np.mean(mae_scores)
avg_rmse = np.mean(rmse_scores)

print(f'{bcolors.BOLD}{bcolors.UNDERLINE}Average R-squared:{bcolors.ENDC} {bcolors.OKGREEN}{avg_r2:.4f}{bcolors.ENDC}')
print(f'{bcolors.BOLD}{bcolors.UNDERLINE}Average Mean Absolute Error:{bcolors.ENDC} {bcolors.OKGREEN}{avg_mae:.4f}{bcolors.ENDC}')
print(f'{bcolors.BOLD}{bcolors.UNDERLINE}Average Root Mean Squared Error:{bcolors.ENDC} {bcolors.OKGREEN}{avg_rmse:.4f}{bcolors.ENDC}')
```

<p align="justify">
    K-fold cross-validation was employed to validate the performance of the linear regression model. The dataset was divided into five equally sized folds using the KFold method from the scikit-learn library. Performance metrics, including R-squared, Mean Absolute Error (MAE), and Root Mean Squared Error (RMSE), were computed for each iteration to evaluate how well the model generalized to unseen data. By averaging these metrics across all folds, a robust estimate of the model’s predictive accuracy and reliability was obtained, minimizing the risk of overfitting and ensuring a more comprehensive assessment of its performance.
</p>

<h2><center>Model Evaluation</center></h2>

<h3>Performance Metrics</h3>

```python
# To view this code, click the three dots -->

# Method to calculate R-Squared
def calculate_r_squared(y1, predictions):
    ss_total = np.sum((y1 - np.mean(y1)) ** 2)
    ss_residual = np.sum((y1 - predictions) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    return r_squared

# Method to calculate MAE
def calculate_mae(y1, predictions):
    mae = np.mean(np.abs(y1 - predictions))
    return mae

# Method to calculate Root Mean Squared Error (RMSE)
def calculate_rmse(y1, predictions):
    mse = np.mean((y1 - predictions) ** 2)
    rmse = np.sqrt(mse)
    return rmse
```

```python
# To view this code, click the three dots -->

# Using the optimized parameters to make predictions
predictions = x1 * m + b

# Use methods to evaluate the model
r_squared = calculate_r_squared(y1, predictions)
mae = calculate_mae(y1, predictions)
rmse = calculate_rmse(y1, predictions)

print(f"{bcolors.BOLD}{bcolors.UNDERLINE}R-squared:{bcolors.ENDC} {bcolors.OKGREEN}{r_squared:.4f}{bcolors.ENDC}")
print(f"{bcolors.BOLD}{bcolors.UNDERLINE}Mean Absolute Error (MAE):{bcolors.ENDC} {bcolors.OKGREEN}{mae:.4f}{bcolors.ENDC}")
print(f"{bcolors.BOLD}{bcolors.UNDERLINE}Root Mean Squared Error (RMSE):{bcolors.ENDC} {bcolors.OKGREEN}{rmse:.4f}{bcolors.ENDC}")
```

<p align="justify">
    The performance metrics calculated and displayed above provide a comprehensive understanding of its accuracy and reliability. An R-squared value of 0.979 indicates that 97.9% of the variance in the ice cream truck revenue can be explained by the temperature, showcasing a very strong linear relationship between the two variables. The Mean Absolute Error (MAE) of 19.5859 signifies that, on average, the model's predictions deviate from the actual revenue by about $19.59, offering a straightforward measure of prediction accuracy. Meanwhile, the Root Mean Squared Error (RMSE) of 24.9619 provides a similar indication of prediction error but gives more weight to larger errors, averaging the squared differences and then taking the square root. Together, these metrics demonstrate that the model not only fits the data well but also makes precise predictions, with minimal average errors, thereby validating its effectiveness in predicting ice cream truck revenue based on temperature.
</p>
