<h3>Scatter Plot</h3>

```python
# Set x and y values corresponding to temperature and profit columns
x = DataFrame(data, columns=['Temperature'])
y = DataFrame(data, columns=['Ice Cream Profits'])

#Create floating point list of values for each column
temperature = data['Temperature'].tolist()
revenue = data['Ice Cream Profits'].tolist()

# Define and display scatter plot
plt.figure(figsize=(10,6))
plt.scatter(x,y, alpha=0.3)
plt.title('Ice Cream Revenue by Temperature')
plt.xlabel('Temperature in Celsius')
plt.ylabel('Ice Cream Revenue $')
plt.ylim(10, 1000)
plt.xlim(0, 45.0)
plt.show()
```

<h2><center>Correlation Analysis</center></h2>

<h3>Pearson Correlation Coefficient</h3>

```python
# Calculate and display Pearson Correlation Coefficient
corr_coefficient, p_value = pearsonr(temperature, revenue)
print(f"Pearson correlation coefficient: {corr_coefficient:.2f}")
```
