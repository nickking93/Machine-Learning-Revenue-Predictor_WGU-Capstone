<h2><center>Model Training</center></h2>

<h3>Linear Regression</h3>

We will employ a linear regression model to predict the revenue of Frosty Forecasts Ice Cream Co. based on the same data being displayed in the charts above. Linear regression aims to establish a linear relationship between temperature and the revenue, and will model this relationship as a straight line. The equation of the line is represented as $y = mx + b$, where $y$ is the predicted revenue, $x$ is the temperature, $m$ is the slope, and $b$ is the y-intercept. 

```python
# To view this code, click the three dots -->

# Linear regression
regression = LinearRegression()
regression.fit(x, y)

# Plot the line
plt.figure(figsize=(10,6))
plt.scatter(x,y, alpha=0.3)
plt.plot(x, regression.predict(x), color='red', linewidth=4)

plt.title('Ice Cream Revenue by Temperature')
plt.xlabel('Temperature in Celsius')
plt.ylabel('Ice Cream Revenue $')
plt.ylim(10, 1000)
plt.xlim(0, 45.0)
plt.show()
```

The chart above draws the the regression line on the previously shown scatterplot and illustrates the relationship between temperature and revenue. While the scatterplot displays individual data points with actual values, the overlaid regression line represents the model’s predicted revenue based on temperature, clearly showing the linear trend the model has identified. This visualization not only helps in understanding the strength and direction of the relationship but also provides a quick assessment of how well the linear regression model fits the data. By comparing the data points to the regression line, we can easily gauge the model's accuracy.

<h3>Gradient Descent</h3>

To optimize the parameters $m$ and $b$, we use gradient descent, an iterative optimization algorithm. Gradient descent minimizes the cost function, which measures the difference between the actual and predicted revenues. By iteratively adjusting $m$ and $b$ in the direction that reduces the cost function, gradient descent efficiently converges to the optimal values that best fit the historical data, thus improving the accuracy of the revenue predictions.

```python
# To view this code, click the three dots -->

# Convert lists to NumPy arrays
x1 = np.array(temperature)
y1 = np.array(revenue)
m1 = regression.coef_[0][0]
b1 = regression.intercept_[0]
# Add a column of ones to x1 to include the intercept term
x1_b = np.c_[x1, np.ones_like(x1)]

# Initialize parameters
m = 21  # Slope (coefficient for temperature)
b = 44  # Intercept
learning_rate = 0.01
precision = .00000001  # Precision for convergence
max_iterations = 5000000  # Maximum number of 

# Function to compute the cost (Mean Squared Error)
def compute_cost(x1, y1, m, b):
    predictions = x1 @ np.array([m, b])
    errors = predictions - y1
    return np.mean(errors ** 2)
    
# Set up the plot
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1, color='blue', label='Data Points', alpha=0.3)
line, = plt.plot(x1, x1 * m + b, color='red', label='Regression Line')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.title('Gradient Descent Optimization of Linear Regression')
plt.legend()

# Gradient Descent with Convergence Criteria
previous_cost = float('inf')  # Initialize with a large 

for iteration in range(max_iterations):
    # Compute predictions
    predictions = x1_b @ np.array([m, b])
    
    # Compute gradients
    error = predictions - y1
    gradient_m = (2 / len(y1)) * np.sum(error * x1)
    gradient_b = (2 / len(y1)) * np.sum(error)
    
    # Update parameters
    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b
    
    # Compute the current cost
    current_cost = compute_cost(x1_b, y1, m, b)
    stepsize = abs(previous_cost - current_cost)
    
    # Check for convergence
    if stepsize < precision:
        print(f"Converged after {iteration+1} iterations.")
        break
    else:
        break
    
    # Update the previous cost
    previous_cost = current_cost
    
    # Update the regression line in the plot
    line.set_ydata(x1 * m + b)
    plt.pause(0.01)  # Pause to update the plot

print(f"Starting slope (m): {m1} ")
print(f"Optimized slope (m): {m} ")
print('')
print(f"Starting intercept (b): {b1}")
print(f"Optimized intercept (b): {b}")

# Using the optimized parameters to make predictions
def predict(x1):
    return m * x1 + b
```

Despite the small changes observed in the optimized values of $m$ (slope) and $b$ (intercept), the process of applying gradient descent and visualizing the updates remains highly instructive and valuable. The initial regression line already provided extremely accurate predictions, indicating a strong fit to the data. However, the gradient descent optimization allowed for fine-tuning of the parameters, ensuring that even minor inaccuracies were minimized. The chart was updated to reflect these small adjustments and visually demonstrates the robustness and precision of the model. This iterative approach reinforces the reliability of the model's predictions and highlights the power of gradient descent in achieving optimal parameter values, even when starting from a nearly optimal state. Thus, the exercise underscores the importance of gradient descent in model optimization and provides a clear, visual confirmation of its effectiveness.
